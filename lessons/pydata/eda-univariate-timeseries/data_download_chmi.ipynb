```python
import requests
import pandas as pd
import re
from pathlib import Path
import time

def list_and_download_chmi_files_simple():
    """
    List all files from CHMI temperature data directory and download those
    containing Ruzyně WSI code (11518) - simplified version using regex
    """
    # Base URL for CHMI temperature data
    base_url = "https://opendata.chmi.cz/meteorology/climate/historical_csv/data/daily/temperature/"

    # Get Ruzyně WSI codes from metadata
    print("Getting Ruzyně WSI codes from metadata...")
    meta1 = pd.read_csv("https://opendata.chmi.cz/meteorology/climate/historical_csv/metadata/meta1.csv")
    meta_ruzyne = meta1.loc[meta1["FULL_NAME"].str.contains("Ruzyně")]
    ruzyne_wsi_codes = meta_ruzyne["WSI"].unique()

    print(f"Found Ruzyně WSI codes: {ruzyne_wsi_codes}")

    # Extract the numeric part from WSI codes (e.g., "0-20000-0-11518" -> "11518")
    ruzyne_ids = [wsi.split('-')[-1] for wsi in ruzyne_wsi_codes]
    print(f"Looking for files containing IDs: {ruzyne_ids}")

    # Get directory listing
    print(f"Fetching directory listing from {base_url}...")
    response = requests.get(base_url)
    response.raise_for_status()

    # Use regex to find CSV file links in the HTML
    html_content = response.text
    csv_pattern = r'<a href="([^"]*\.csv)"'
    csv_matches = re.findall(csv_pattern, html_content)

    print(f"Found {len(csv_matches)} total CSV files")

    # Filter files for Ruzyně stations
    ruzyne_files = []
    for file in csv_matches:
        for ruzyne_id in ruzyne_ids:
            if ruzyne_id in file:
                ruzyne_files.append(file)
                break

    print(f"Found {len(ruzyne_files)} files for Ruzyně stations:")
    for file in sorted(ruzyne_files):
        print(f"  - {file}")

    # Download Ruzyně files
    downloaded_files = []
    for file in ruzyne_files:
        file_url = base_url + file
        local_path = Path(file)

        print(f"Downloading {file}...")
        try:
            file_response = requests.get(file_url)
            file_response.raise_for_status()

            with open(local_path, 'wb') as f:
                f.write(file_response.content)

            downloaded_files.append(str(local_path))
            print(f"  ✓ Saved to {local_path}")

            # Be nice to the server
            time.sleep(0.5)

        except Exception as e:
            print(f"  ✗ Error downloading {file}: {e}")

    print(f"\nDownload complete! Downloaded {len(downloaded_files)} files:")
    for file in downloaded_files:
        print(f"  - {file}")

    return downloaded_files, csv_matches, ruzyne_files

# Run the simplified function
downloaded_files, all_files, ruzyne_files = list_and_download_chmi_files_simple()
```
